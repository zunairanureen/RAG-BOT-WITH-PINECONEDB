```python
from langchain_community.document_loaders import PyPDFLoader, PyPDFDirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_pdf_files(data):
    loader = PyPDFDirectoryLoader(data,
                                  glob="*.pdf")
    documents=docs = loader.load()
    return documents
```

```python
extracted_data=load_pdf_files(data="data/")
```

```python
extracted_data
```

```python
def text_split(extracted_data):
    text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size=400,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)
    text_chunks=text_splitter.split_documents(extracted_data)
    return text_chunks
```

```python
chunks = text_split(extracted_data=extracted_data)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

def load_huggingfaceembeddings(chunks):
    # Load Hugging Face embeddings model
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    
    # Extract the page content from the Document objects before embedding
    texts = [chunk.page_content for chunk in chunks]  
    
    # Create embeddings for the chunks
    embeddings_result = embeddings.embed_documents(texts)  # Pass the extracted texts
    
    return embeddings_result
```
